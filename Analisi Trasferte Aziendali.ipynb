```python
import pandas as pd
import random 
from datetime import datetime, timedelta


employees = [
    "Marco Rossi",
    "Laura Bianchi",
    "Andrea Verdi",
    "Giulia Ferrari",
    "Luca Moretti"
]

cities = {
    "Verona": 45,
    "Milano": 160,
    "Vicenza": 45,
    "Trento": 100,
    "Bologna": 115,
    "Padova": 90,
    "Brescia": 75,
    "Mantova": 40
}

clients = [
    "Alfa SRL",
    "Beta SPA",
    "Gamma Tech",
    "Delta Medical",
    "Omega Group",
    "Sigma Consulting"
]

cars = {
    "Fiat Panda": 18,
    "Volkswagen Golf": 16,
    "Audi A3": 14,
    "Toyota Yaris": 20
}



rows = []

start_date = datetime(2024, 1, 1)

for i in range(300):

    trip_date = start_date + timedelta(days=random.randint(0, 364))

    employee = random.choice(employees)

    departure = "Verona"

    destination = random.choice(list(cities.keys()))

    while destination == departure:
        destination = random.choice(list(cities.keys()))

    client = random.choice(clients)

    car = random.choice(list(cars.keys()))

    km_per_liter = cars[car]

    distance_km = cities[destination] + random.randint(-10, 10)

    fuel_price = round(random.uniform(1.75, 2.10), 2)

    fuel_used = distance_km / km_per_liter

    travel_cost = round(fuel_used * fuel_price, 2)

    rows.append([
        trip_date,
        employee,
        client,
        departure,
        destination,
        car,
        km_per_liter,
        distance_km,
        fuel_price,
        travel_cost
    ])


    columns = [
    "Date",
    "Employee",
    "Client",
    "Departure",
    "Destination",
    "Car",
    "Km_per_Liter",
    "Distance_KM",
    "Fuel_Price",
    "Travel_Cost"
]

travel_df = pd.DataFrame(rows, columns=columns)

travel_df.to_csv("business_travel_dataset.csv", index=False)

print("Dataset creato correttamente!")

```
